```python
import os
import shutil

data_dir = '/content/final_dataset/agri_data/data'
images_dir = os.path.join(data_dir, 'images')
labels_dir = os.path.join(data_dir, 'labels')

os.makedirs(images_dir, exist_ok=True)
os.makedirs(labels_dir, exist_ok=True)

for filename in os.listdir(data_dir):
    if filename.endswith(('.jpeg', '.jpg', '.png')):
        shutil.move(os.path.join(data_dir, filename), images_dir)
    elif filename.endswith('.txt'):
        shutil.move(os.path.join(data_dir, filename), labels_dir)

print("Files moved to 'images' and 'labels' directories.")
```

```python
data_yaml = """
path: /content/final_dataset/agri_data/data

train: images
val: images

names:
  0: crop
  1: weed
"""

with open('/content/data.yaml', 'w') as f:
    f.write(data_yaml)

print("data.yaml updated successfully.")
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

model.train(
    data='/content/data.yaml',
    epochs=25,
    imgsz=512,
    batch=8
)
```

```python
from ultralytics import YOLO

model = YOLO('/content/runs/detect/train-2/weights/best.pt')

results = model.predict(
    source='/content/Screenshot 2026-05-14 000927.png',
    save=True,
    conf=0.5
)

print("Detection Completed")
```

```python
import cv2
import matplotlib.pyplot as plt
import os # Import os for path existence check

# Assuming 'results' is available from the previous cell's execution
# It's a list, so we take the first element (for the single image prediction)
predicted_image_path = results[0].path if results and len(results) > 0 else None

if predicted_image_path and os.path.exists(predicted_image_path):
    img = cv2.imread(predicted_image_path)
    # Check if the image was loaded successfully before converting color
    if img is not None:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        plt.figure(figsize=(10,10))
        plt.imshow(img)
        plt.axis('off')
        plt.show()
    else:
        print(f"Error: OpenCV could not read the image file at {predicted_image_path}")
else:
    print(f"Error: Predicted image file not found at {predicted_image_path}")
```
